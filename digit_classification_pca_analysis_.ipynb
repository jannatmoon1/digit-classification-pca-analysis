```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
digits.data[0]
```

```python
df= pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
x=df
y=digits.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scalled=scaler.fit_transform(x)
x_scalled[0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scalled,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
x_pca.shape
```

```python
x_pca[0]
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
pca = PCA(n_components=2)
x1_pca = pca.fit_transform(x)
x1_pca.shape
```

```python
x1_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(x1_pca, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
